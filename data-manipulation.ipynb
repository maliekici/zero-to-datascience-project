```python
df.drop_duplicates() #column tamamen aynı olanları siler
```

```python
#columnlar bazında aynı olanları düşürmek istersek
df.drop_duplicates(subset=["Sex"])
#sadece 2 tane farklı olduğu için aynı olanların hepsini sildi ve 2 tanesi kaldı
```

```python
df.drop_duplicates(subset=["Sex"], keep='last')
#bi üstteki kodda 0 ve 2. indexteki kayıtları tuttu keep parametresini last
#olarak ayarlarsak en sondakini tutar.
```

```python
employee_data = {'employee_id' : ['1','2','3','4'],
 'name': ['mert cobanoc', 'fethi takyaygil', 'segaf blackwater',
 'tim hortam']}
sales_data = {'employee_id' : ['3','4','5','6'],
 'total sales': [23456, 2512, 2345, 1455 ]}
```

```python
employee = pd.DataFrame(employee_data)
sales = pd.DataFrame(sales_data)
```

```python
pd.merge(employee, sales, on='employee_id', how='right')
#how'a değer vermez ise sadece 3,4 ü yani iki taraf içinde dataları olan kısımı birleştirir
#outher bütün satayı NaN değer olsun olmasın birleştirir
#left soldaki yani employee_data üzerinde hepsi var olan üzerinden birleştirir
#right letfin tam tersi
```
