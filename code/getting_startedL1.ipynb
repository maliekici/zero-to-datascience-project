```python
image = cv2.imread('./img_data/1.jpg')
# cv2.imshow("image", image) #if you are working with the IDE you downloaded to your computer, this command will probably work for you.
```

```python
#displaying images
from matplotlib import pyplot as plt
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)) 
plt.show()
```

```python
def image_show(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
#lets call our function
image3 = cv2.imread('./img_data/3.jpg')
image_show("display", image3)
```

```python
#saving image
cv2.imwrite("outputimg3.jpg", image3) #or .png
```
